#### Data description:

   * "id" : App ID

   * "track_name": App Name

   * "size_bytes": Size (in Bytes)

   * "currency": Currency Type

   * "price": Price amount

   * "ratingcounttot": User Rating counts (for all version)

   * "ratingcountver": User Rating counts (for current version)

   * "user_rating" : Average User Rating value (for all version)

   * "userratingver": Average User Rating value (for current version)

   * "ver" : Latest version code

   * "cont_rating": Content Rating

   * "prime_genre": Primary Genre

   * "sup_devices.num": Number of supporting devices

   * "ipadSc_urls.num": Number of screenshots showed for display

   * "lang.num": Number of supported languages

   * "vpp_lic": Vpp Device Based Licensing Enabled

## Create a new column 'payment_type' and categorize all the applications into 'free' or 'paid'

```python
# We will be using the apply function here to get results

# Custom function to classify an app as free or paid
def free_paid(x):
    if x > 0.00:
        return 'paid'
    else:
        return 'free'

# creating a new column as per our requirement
apple['payment_type'] = apple['price'].apply(lambda x : free_paid(x))

apple.head()
```

## Number of free apps should be extremly high when compared to paid apps. Lets do a reality check.


```python
# A countplot to check the number of free and paid apps
sns.countplot(x=apple.payment_type,data=apple)
```

## In which price range do most of the paid apps below 50 dollars fall?

```python
# creating a subset for only the paid apps
df = apple[apple['price'] != 0]

# Setting plot details
plt.figure(figsize = (15,10))
plt.style.use("fivethirtyeight")

# creating a subset of the data below 50 dollars and plotting the data
dfa = df[df['price'] < 49.99]

plt.xlabel('Price($)')

dfa.price.plot(kind = 'hist',log = True)
```

#### Insights
Most apps with maxx price < 49.99 lie in the $(0-5) range, as the price of the apps increase the frequency count decrease, as is evident from the histogram.

As the price increases the frequency count decreases which indicates that ideally price should be lower or equal to $30, after that, the Frequency Count decreases exponentially

#### The number of free apps are more but not drastically as per our initial belief

## Find the top 5 "prime_genre' in terms of count. Are the paid apps for these genres good enough?

```python
# Finding the top 5 genres
top_genre = apple.prime_genre.value_counts().sort_values(ascending = False)
top_genre.head()
```

```python
# creating a list of top 5 genres to create a subset of the same
top_genre = top_genre[:5].index
top_genre_df = apple[apple['prime_genre'].isin(top_genre)]
```

```python
# plotting a violin plot to understand the ratings
plt.figure(figsize = (20,15))
plt.ylim(0,5)
sns.violinplot(data = top_genre_df, y = 'user_rating', x = 'prime_genre', hue = 'payment_type', split = True, scale = 'count', palette= ['#44a8c5','#5fc497'] )
plt.style.use("fast")
plt.xlabel('Categories')
plt.ylabel('User Rating')
plt.title('User Rating Distribution')
plt.show()
```

#### Insight:
* For Games and Photo & Video Genres, the median rating of the apps is approximately 4.5, which suggests that they are better

### Create a new column "size_mb" converting bytes to megabytes. For the applications below 50 dollars, does the price of the app increases with the increase in its size?

```python
# Converting bytes to megabytes
def size(x):
    mb = x/1000000
    return mb
    
top_genre_df['size_mb'] = apple.size_bytes.apply(lambda x : size(x))
top_genre_df.head(5)
```

```python
# creating a scatter plot to get our answer
plt.figure(figsize = (15,10))
paid_apps_desc = top_genre_df[top_genre_df['price'] <50]

sns.lmplot(data = paid_apps_desc, x = 'size_mb', y = 'price', col = 'prime_genre',col_wrap= 3,aspect= 1.5
           ,scatter = True,fit_reg = False, hue  = 'prime_genre', legend_out = True,
           palette = ['#c973d0' , '#8e52b8' , '#4a4f92', '#4a73ab' , '#649ca1'] )
plt.show()
```

* The relation between Price and Genere suggest that there exists no relation between app size and price

## Can we create a recommendation column based on "user_rating" and "rating_count_tot"?

```python
def recommendation(total_reviews,rating):
# creating an empty list to store the recommendation
    recommendation = []
# Iterating through all the rating and review count values, using the if else statement and categorizing the app
    for review_count,avg_rating in zip(apple['rating_count_tot'],apple['user_rating']):
        if review_count > total_reviews and avg_rating > rating:
            recommendation.append('yes')
        else:
            recommendation.append('no')
# Creating a new recommendation column
    apple['recommendation'] = recommendation
```

```python
recommendation(20000,3.5)
```
